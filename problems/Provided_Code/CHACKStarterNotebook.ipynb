## Load abundance data and outputs/metals data:

Now that we have the CHACKutils module imported, we can use its functions to load the datasets for us. First, load the data from the csv files using the 

```
CHACKutils.dataloader()
```
function. This function has the optional parameter `multiIndex`. If set to `True`, the function will return the full multi-indexed dataframe that contains information on the full taxonomic classification of each bacteria. __If you are doing project one, you probably do not want to include `multiIndex=True`__.

Watch out - **`dataloader()` will return 2 dataframes**. The first contains the bacterial abundance data, and the second contains the outputs and metals data.



````python
species_data, outputs_and_metals = CHACKutils.dataloader(multiIndex=True)
````

If you are doing Project 2, you may want to consider the bacterial abundance data at different taxonomic levels. This is easy to do using the `getTaxonomicLevelData` function. It will take the species data dataframe returned from the dataloader function (called using the `multiIndex=True` option), and return a dataframe of the abundance data at your desired taxonomic level. We reccomend running the abundance data through here even if you want to look at the species data (ie, call the function with the 'Species' argument, as the DataFrame returned from this function will be easier to work with than the multiIndexed dataframe it takes in.

````python
CHACKutils.getTaxonomicLevelData('Family', species_data)
````
